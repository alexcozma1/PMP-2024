```python
from pgmpy.models import MarkovNetwork
import networkx as nx

model = MarkovNetwork([('A', 'B'), ('B', 'C'), ('A', 'D'), ('D', 'E')])

pos = nx.circular_layout(model)
nx.draw(model, with_labels=True, pos=pos, alpha=0.5, node_size=2000)
```
