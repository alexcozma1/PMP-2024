```python
import pymc as pm
import numpy as np
import arviz as az

observatii_aruncari = ['s', 's', 'b', 'b', 's', 's', 's', 's', 'b', 's']
nr_steme = sum(1 for _ in observatii_aruncari if _ == 's')
nr_aruncari = len(observatii_aruncari)

with pm.Model() as model:
    p = pm.Beta("p", alpha = 1, beta = 1)

    verosimilitate = pm.Binomial("verosimilitate", p = p, n = nr_aruncari, observed = nr_steme)

    trace = pm.sample(2000, tune=1000, return_inferencedata=True)

az.plot_posterior(trace, var_names=["p"], hdi_prob=0.95)
az.summary(trace, var_names=["p"])
```
